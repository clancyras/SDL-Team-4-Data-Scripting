```python
import pandas as pd
from scipy.stats import linregress

# Set Pandas options to display all rows and columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Adjust to terminal width
pd.set_option('display.max_colwidth', None)  # No truncation of column values

def calculate_r_squared(csv_file):
    # Load CSV file
    df = pd.read_csv(csv_file)
    
    # include only numerical columns
    numeric_df = df.select_dtypes(include=['number'])
    
    # Drop rows with missing values
    numeric_df = numeric_df.dropna()

    # column names
    columns = numeric_df.columns
    r_squared_results = {}

    # Compute R-squared values for each pair in df
    for i in range(len(columns)):
        for j in range(i+1, len(columns)):  # Avoid duplicate pairs
            col_x = columns[i]
            col_y = columns[j]

            try:
                # Perform linear regression
                slope, intercept, r_value, p_value, std_err = linregress(numeric_df[col_x], numeric_df[col_y])
                
                # Compute R-squared
                r_squared = r_value ** 2
                
                # Store in dictionary
                r_squared_results[(col_x, col_y)] = r_squared
            
            except Exception as e:
                print(f"Skipping ({col_x}, {col_y}) due to error: {e}")

    # Convert results into a DataFrame for better visualization
    r_squared_df = pd.DataFrame(list(r_squared_results.items()), columns=["Column Pair", "R-Squared"])
    
    return r_squared_df

# Usage
csv_file = "datasets/cleaned_dataset.csv"  # Replace with your actual file path
r_squared_data = calculate_r_squared(csv_file)
r_squared_data = r_squared_data.sort_values(by="R-Squared", ascending=False)
print(r_squared_data)
```
