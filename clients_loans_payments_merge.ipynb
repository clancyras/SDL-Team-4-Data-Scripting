```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv("datasets/clients_loans_payments/clients.csv")
df2 = pd.read_csv("datasets/clients_loans_payments/loans.csv")
df3 = pd.read_csv("datasets/clients_loans_payments/payments.csv")

# Merge on the shared column (replace 'SharedColumn' with the actual column name)
df_merged = df1.merge(df2, on="client_id", how="outer").merge(df3, on="loan_id", how="outer")

# Save the merged DataFrame to a new CSV file
df_merged.to_csv("datasets/clients_loans_payments/merged_output.csv", index=False)

# Display the first few rows
print(df_merged.head())
```
