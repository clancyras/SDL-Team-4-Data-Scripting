```python
import csv
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dataset_loc = "datasets/Loan_status_2007-2020Q3.gzip"
y_name = 'total_credit_limit'
x_name = 'loan_amount'
is_color = False
color_name = 'fico_range_low'
```

```python

import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv(dataset_loc)  # Replace with your CSV file
for i, col in enumerate(df.columns):
    print(col, i)
# Choose data columns
x = df[x_name]  # Replace with appropriate column name

y = df[y_name]  # Replace with appropriate column name
print(f"Correlation: {x.corr(y)}")
if is_color:
    c = (df[color_name] - df[color_name].min()) / \
                (df[color_name].max() - df[color_name].min())
# Create the dot plot
plt.figure(figsize=(10, 5))

if is_color:
    scatter = plt.scatter(x, y, c=c, cmap="viridis", s=4, alpha=0.2, edgecolors=None)
    plt.colorbar(scatter, label=color_name)
else:
    scatter = plt.scatter(x, y, color='blue', s=5, alpha=0.2)
# Labels and title
plt.xlabel(x_name)
plt.ylabel(y_name)
if is_color:
    plt.title(f"{x_name}, {y_name}, {color_name}")
else:
    plt.title(f"{x_name}, {y_name}")

plt.xticks(rotation=45)  # Rotate x-axis labels if needed

# Show the plot
plt.show()

```
