```python
import pandas as pd

df = pd.read_csv("datasets/lending_club/Loan_status_2007-2020Q3.gzip")

#Only include specified columnes
df_cleaned = pd.DataFrame()

# Integer columns
df_cleaned['loan_amnt'] = pd.to_numeric(df['loan_amnt'], errors='coerce').fillna(0).astype(int)
df_cleaned['term'] = df['term'].str.extract('(\d+)').astype(float).fillna(0).astype(int)  # '36 months' -> 36
df_cleaned['annual_inc'] = pd.to_numeric(df['annual_inc'], errors='coerce').fillna(0).astype(int)
df_cleaned['fico_range_low'] = pd.to_numeric(df['fico_range_low'], errors='coerce').fillna(0).astype(int)
df_cleaned['fico_range_high'] = pd.to_numeric(df['fico_range_high'], errors='coerce').fillna(0).astype(int)
df_cleaned['emp_length'] = df['emp_length'].str.extract('(\d+)').astype(float).fillna(0).astype(int)
df_cleaned['delinq_2yrs'] = pd.to_numeric(df['delinq_2yrs'], errors='coerce').fillna(0).astype(int)
df_cleaned['inq_last_6mths'] = pd.to_numeric(df['inq_last_6mths'], errors='coerce').fillna(0).astype(int)
df_cleaned['mths_since_last_delinq'] = pd.to_numeric(df['mths_since_last_delinq'], errors='coerce').fillna(0).astype(int)

df_cleaned['int_rate'] = df['int_rate'].str.rstrip('%').astype(float) / 100 # Strip % and convert to float

# Convert float columns to two decimal places
df_cleaned['installment'] = df['installment'].astype(float).round(2)
df_cleaned['dti'] = df['dti'].astype(float).round(2)
df_cleaned['total_pymnt'] = df['total_pymnt'].astype(float).round(2)

# Convert dates to YYYY/MM
df_cleaned['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y', errors='coerce').dt.strftime('%Y/%m') 
df_cleaned['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%Y', errors='coerce').dt.strftime('%Y/%m')

# Categories
df_cleaned['loan_status'] = df['loan_status'].astype('category')
df_cleaned['purpose'] = df['purpose'].astype('category')
df_cleaned['home_ownership'] = df['home_ownership'].astype('category')

# Upercase
df_cleaned['addr_state'] = df['addr_state'].str.upper()

df_cleaned.to_csv("datasets/cleaned_dataset.csv", index=False)

```
